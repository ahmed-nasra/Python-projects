### Distribution of Exam Scores

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Exam_Score'], bins=20, kde=True)
plt.title('Distribution of Exam Scores')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.show()
```

### Correlation Analysis

```python
# Select numerical columns
numerical_cols = df.select_dtypes(include=['int64']).columns

# Compute correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Variables')
plt.show()

# Focus on correlations with Exam_Score
print("\nCorrelations with Exam_Score:")
print(correlation_matrix['Exam_Score'].sort_values(ascending=False))
```

### Categorical Variables vs. Exam Score

```python
# List of key categorical columns to analyze
key_categorical_cols = ['Parental_Involvement', 'Access_to_Resources', 'Extracurricular_Activities',
                       'Motivation_Level', 'Internet_Access', 'Family_Income', 'Teacher_Quality',
                       'School_Type', 'Peer_Influence', 'Learning_Disabilities']

# Plot box plots for each categorical variable
plt.figure(figsize=(15, 20))
for i, col in enumerate(key_categorical_cols, 1):
    plt.subplot(5, 2, i)
    sns.boxplot(x=col, y='Exam_Score', data=df)
    plt.title(f'Exam Score by {col}')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Key Numerical Variables vs. Exam Score

```python
# Scatter plots for key numerical variables
plt.figure(figsize=(15, 5))

# Hours_Studied vs Exam_Score
plt.subplot(1, 3, 1)
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=df)
plt.title('Hours Studied vs Exam Score')

# Attendance vs Exam_Score
plt.subplot(1, 3, 2)
sns.scatterplot(x='Attendance', y='Exam_Score', data=df)
plt.title('Attendance vs Exam Score')

# Previous_Scores vs Exam_Score
plt.subplot(1, 3, 3)
sns.scatterplot(x='Previous_Scores', y='Exam_Score', data=df)
plt.title('Previous Scores vs Exam Score')

plt.tight_layout()
plt.show()
```

### Bar Plot of Average Exam Score by School Type and Gender

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='School_Type', y='Exam_Score', hue='Gender', data=df, ci=None, palette='Set2')
plt.title('Average Exam Score by School Type and Gender')
plt.xlabel('School Type')
plt.ylabel('Average Exam Score')
plt.legend(title='Gender')
plt.show()
```

### Box Plot of Exam Score by Teacher Quality and Peer Influence

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Teacher_Quality', y='Exam_Score', hue='Peer_Influence', data=df, palette='Set3')
plt.title('Exam Score by Teacher Quality and Peer Influence')
plt.xlabel('Teacher Quality')
plt.ylabel('Exam Score')
plt.legend(title='Peer Influence')
plt.show()
```

### Heatmap of Exam Score by Hours Studied and Attendance

```python
# Bin Hours_Studied and Attendance
df['Hours_Studied_Bin'] = pd.cut(df['Hours_Studied'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
df['Attendance_Bin'] = pd.cut(df['Attendance'], bins=5, labels=['60-68%', '68-76%', '76-84%', '84-92%', '92-100%'])

# Create pivot table for heatmap
pivot_table = df.pivot_table(values='Exam_Score', index='Hours_Studied_Bin', columns='Attendance_Bin', aggfunc='mean')

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='.1f')
plt.title('Average Exam Score by Hours Studied and Attendance')
plt.xlabel('Attendance')
plt.ylabel('Hours Studied')
plt.show()
```

### Count Plot of Learning Disabilities by Family Income

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Family_Income', hue='Learning_Disabilities', data=df, palette='Pastel1')
plt.title('Count of Learning Disabilities by Family Income')
plt.xlabel('Family Income')
plt.ylabel('Count')
plt.legend(title='Learning Disabilities')
plt.show()
```

### Facet Grid of Exam Score by Motivation Level and Gender

```python
g = sns.FacetGrid(df, col='Motivation_Level', row='Gender', height=4, aspect=1.2)
g.map(sns.histplot, 'Exam_Score', bins=20)
g.set_titles('{col_name} Motivation | {row_name}')
g.set_axis_labels('Exam Score', 'Count')
plt.suptitle('Exam Score Distribution by Motivation Level and Gender', y=1.05)
plt.show()
```

### Line Plot of Average Exam Score by Sleep Hours and Physical Activity

```python
# Group by Sleep_Hours and Physical_Activity, calculate mean Exam_Score
grouped = df.groupby(['Sleep_Hours', 'Physical_Activity'])['Exam_Score'].mean().unstack()

# Plot line plot
plt.figure(figsize=(10, 6))
grouped.plot(kind='line', marker='o')
plt.title('Average Exam Score by Sleep Hours and Physical Activity')
plt.xlabel('Sleep Hours')
plt.ylabel('Average Exam Score')
plt.legend(title='Physical Activity (Hours/Week)', loc='best')
plt.show()
```
